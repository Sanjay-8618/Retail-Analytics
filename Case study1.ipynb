```python
#6. Which City code has the maximum customers and what was the percentage of customers from 
#that city?

customer_final.groupby(by = "city_code").sum()
cust = customer_final['city_code'].value_counts()
city = cust.idxmax()
city
```

```python
#6. Which City code has the maximum customers and what was the percentage of customers from 
#that city?

total_customers = len(customer_final)

percentage_customers = (cust[city] / total_customers) * 100
percentage_customers
```

```python
#8. What was the total amount earned from the "Electronics" and "Clothing" categories from 
#Flagship Stores?

data = customer_final[customer_final['Store_type'].isin(['Flagship store'])]
data = data[data['prod_cat'].isin(['Electronics','Clothing'])]
data['total_amt'].sum()
```

```python
#9. What was the total amount earned from "Male" customers under the "Electronics" category?
f_data = customer_final[(customer_final["Gender"]=="M") & (customer_final['prod_cat']=='Electronics')]
f_data['total_amt'].sum()
```

```python
#10. How many customers have more than 10 unique transactions, after removing all transactions 
#which have any negative amounts?

data = customer_final[customer_final['total_amt']>0]
data = data.groupby(by = 'cust_id')['transaction_id'].nunique()
len(data[data>10])
```

```python
#11. For all customers aged between 25 - 35, find out:
#   a. What was the total amount spent for “Electronics” and “Books” product categories?

customer_final['DOB'] = pd.to_datetime(customer_final['DOB'], format='%d-%m-%Y')
# Calculate the age of each customer
customer_final['Age'] = (pd.to_datetime('today') - customer_final['DOB']).astype('<m8[Y]')

data = customer_final[(customer_final['Age'].between(25,35,inclusive = True)) & (customer_final['prod_cat'].isin(['Electronics','Books'])) &(customer_final['total_amt']>0) ]
data['total_amt'].sum()
```
